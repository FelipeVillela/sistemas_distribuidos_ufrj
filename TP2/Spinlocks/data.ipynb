```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(context = 'paper', style='whitegrid', font_scale = 1.4)
```

```python
path = 'results.csv'
data = pd.read_csv(path, names=['N','K','Tempo'])
data
```

```python
fig = plt.figure(figsize=(12,6))
plt.xlabel('K', fontsize=12)
plt.ylabel('Tempo (s)', fontsize=12)
sns.lineplot(
  data = data,
  x = "K",
  y = "Tempo",
  hue = "N",
  linewidth = 3,
  palette = ["red", "orange", "pink"]
)
plt.savefig('plot.png')
plt.show()
```
